```python
#import yfinance as yf
#import pandas as pd
#import matplotlib.pyplot as plt
from my_imports import *
# 테슬라와 제네럴 모터스의 주가 데이터를 가져오기
start_date = '2014-01-01'  # 시작 날짜 설정
end_date = '2024-11-06'    # 종료 날짜 설정

tesla = yf.download('TSLA', start=start_date, end=end_date)['Close']
gm = yf.download('GM', start=start_date, end=end_date)['Close']

data = pd.concat([tesla, gm], axis=1)

    # 결측치 제거
data.dropna(inplace=True)
    # 상관관계 계산
correlation = data.corr().iloc[0, 1]

    # 상관관계 출력
print(f"테슬라와 제네럴 모터스의 주가 상관관계: {correlation:.2f}")

    # 시각화
plt.figure(figsize=(10, 6))
plt.plot(data['TSLA'], label='Tesla', color='blue')
plt.plot(data['GM'], label='General Motors', color='red')
plt.title("Tesla vs General Motors Stock Prices")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```
